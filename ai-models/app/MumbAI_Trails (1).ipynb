```python
import torch
import re
import pandas as pd
from sentence_transformers import util

def recommend_places1(user_prompt, top_k=5, min_score_threshold=0.45):
    """
    Full-proof recommender that handles:
    - Negations: "not crowded", "no beach"
    - Activity intents: "bowling", "cycling", "romantic dinner"
    - Location context: "near Bandra", "in Navi Mumbai"
    - Descriptive adjectives: "quiet", "budget", "luxury", etc.
    - Smart fallbacks for low confidence or no matches
    - Keyword-based boosting
    - Clean explanations for transparency
    """

    print(f"\n🧭 Recommendations for: \"{user_prompt}\"\n")
    user_prompt = user_prompt.lower().strip()

    # --- STEP 1: Extract negations (no / not ___)
    negations = re.findall(r"(?:no|not)\s+([a-zA-Z]+)", user_prompt)

    # --- STEP 2: Detect locations (e.g., near Bandra, in Navi Mumbai)
    loc_match = re.findall(r"(?:in|near|around)\s+([a-zA-Z\s]+)", user_prompt)
    location_filter = [loc.strip() for loc in loc_match]

    # --- STEP 3: Detect mood or tone (quiet, romantic, fun, etc.)
    mood_keywords = {
        "quiet": 1.2, "peaceful": 1.2, "romantic": 1.25, "adventurous": 1.2,
        "fun": 1.2, "budget": 1.15, "luxury": 1.2, "heritage": 1.25,
        "historic": 1.25, "religious": 1.15, "spiritual": 1.15,
        "family": 1.1, "friends": 1.1, "night": 1.1, "morning": 1.05
    }
    boost_weights = {k: v for k, v in mood_keywords.items() if k in user_prompt}

    # --- STEP 4: Detect activity intents ---
    activity_map = {
        "bowling": ["bowling", "game zone", "arcade", "mall"],
        "cycling": ["cycle track", "trail", "park"],
        "trekking": ["fort", "hill", "trek", "mountain"],
        "hiking": ["trail", "forest", "fort", "hill"],
        "museum": ["museum", "gallery", "heritage"],
        "temple": ["temple", "spiritual", "ashram"],
        "dinner": ["restaurant", "cafe", "dining"],
        "date": ["romantic", "restaurant", "view"],
        "shopping": ["mall", "market", "bazaar"],
        "sports": ["stadium", "complex", "ground", "arena"],
        "movie": ["cinema", "theatre", "multiplex"],
        "picnic": ["park", "garden", "lawn"],
    }

    activity_keywords = []
    for act, related in activity_map.items():
        if act in user_prompt:
            activity_keywords.extend(related)

    # --- STEP 5: Apply negation filtering (e.g. remove beaches if "no beach") ---
    filtered_df = df.copy()
    for neg in negations:
        mask = (
            ~filtered_df['category'].str.contains(neg, case=False, na=False) &
            ~filtered_df['subcategory'].str.contains(neg, case=False, na=False) &
            ~filtered_df['description'].str.contains(neg, case=False, na=False)
        )
        filtered_df = filtered_df[mask]

    # --- STEP 6: Filter by location if mentioned ---
    if location_filter:
        loc_mask = pd.Series(False, index=filtered_df.index)
        for loc in location_filter:
            loc_mask |= filtered_df['location'].str.contains(loc, case=False, na=False)
        if loc_mask.any():
            filtered_df = filtered_df[loc_mask]
        else:
            print(f"📍 No exact matches found for {', '.join(location_filter)} — showing broader results.\n")

    if filtered_df.empty:
        print("⚠️ No results after filtering — falling back to full dataset.\n")
        filtered_df = df.copy()

    # --- STEP 7: Compute semantic similarity ---
    user_embedding = model.encode([user_prompt], convert_to_tensor=True)
    filtered_embeddings = place_embeddings[filtered_df.index]
    cos_scores = util.cos_sim(user_embedding, filtered_embeddings)[0]
    scores = cos_scores.clone()

    # --- STEP 8: Apply boosting for tone & activities ---
    matched_keywords = {}
    for i, desc in enumerate(filtered_df['description'].str.lower()):
        matched = []
        for word, weight in boost_weights.items():
            if word in desc:
                scores[i] *= weight
                matched.append(word)
        for word in activity_keywords:
            if word in desc:
                scores[i] *= 1.25
                matched.append(word)
        matched_keywords[i] = matched

    # --- STEP 9: Smart fallback if all low confidence ---
    if scores.max().item() < min_score_threshold:
        print("⚠️ No strong matches found — showing most semantically similar places.\n")

    # --- STEP 10: Sort and display ---
    sorted_indices = torch.argsort(scores, descending=True)
    shown_names = set()
    count = 0

    for idx in sorted_indices.tolist():
        row = filtered_df.iloc[idx]
        score = scores[idx].item()
        name = row.get('name', row.get('subcategory', 'Unknown'))

        if name in shown_names:
            continue
        shown_names.add(name)

        matched = ", ".join(matched_keywords.get(idx, [])) or "semantic similarity"
        print(f"🏙️ {name} ({score:.4f})")
        print(f"{row['description']}")
        print(f"✨ Matched for: {matched}")
        print("-" * 80)

        count += 1
        if count >= top_k:
            break

    if count == 0:
        print("😕 No relevant places found. Try a simpler or clearer query.")
```
